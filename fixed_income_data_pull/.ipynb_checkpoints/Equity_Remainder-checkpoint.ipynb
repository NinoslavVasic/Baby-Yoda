```python
#Read In Fixed Income Prices
equity_rtns = pd.read_csv('equity_returns.csv')
equity_rtns.tail()
```

```python
#Cleaning Up CSV by dropping NAs
equity_clean = equity_rtns.dropna()
equity_clean
```

```python
#New Index Col for Equity Data frame 
equity_clean_2 = equity_clean.set_index('Unnamed: 0')
equity_clean_2.head()
```

```python
#Final Cleaning for Equity DataFrame -> Setting Index to date 
final_equity = equity_clean_2.rename_axis('Date', axis='index')
final_equity
```

```python
#Equity Returns 
equity_returns = final_equity.astype(float).pct_change()
equity_returns
```

```python
#Read In S&P 500 Prices
spy_org = pd.read_csv('spy_returns.csv')
spy_org.head()
```

```python
#Cleaning Up CSV by dropping NAs
spy_clean = spy_org.dropna()
spy_clean.tail()
```

```python
#Drop the final 2 rows 
spy_clean_2 = spy_clean.drop([1257,1258])
spy_clean_2.tail()
```

```python
#New Index Col for SPY Data frame 
spy_clean_3 = spy_clean_2.set_index('date')
spy_clean_3.head()
```

```python
#Cleaning for SPY DataFrame -> Setting Index to date 
spy_clean_4 = spy_clean_3.rename_axis('Date', axis='index')
spy_clean_4.head()
```

```python
#Set Close to SPY for Final DataFrame
spy_final = spy_clean_4.rename(columns = {'close':'SPY'})
spy_final
```

```python
#SPY Returns 
spy_returns = spy_final.astype(float).pct_change()
spy_returns
```

```python
#Concat SPY with Equity DF
eq_rtns_universe = pd.concat([equity_returns, spy_returns],axis = 'columns', join = 'inner')
eq_rtns_universe
```
