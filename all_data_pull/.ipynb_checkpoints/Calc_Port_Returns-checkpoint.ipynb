```python
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
%matplotlib inline
import os
```

```python
blend_opt_weights = [0.22876, 0.20055, 0.31059, 0.2601]
fixed_income_opt_weights = [1]
fixed_income_equal_weights = [0.143, 0.143, 0.143, 0.143, 0.143, 0.143, 0.143]
equity_opt_weights = [0.28993, 0.36465, 0.34542]
equity_equal_weight_weights = [0.833, 0.833, 0.833, 0.833, 0.833, 0.833, 0.833, 0.833, 0.833, 0.833, 0.833, 0.833]
```

```python
#Read In First Equity Dataframe Prices
first_equity_rtns = pd.read_csv('Stock ETF data pull.csv')
first_equity_rtns.head()
```

```python
#Dropping the First Row in the Equity DataFrame
equity_clean = first_equity_rtns.drop([0])
equity_clean
```

```python
#New Index Col for ETF Data frame 
equity_clean_2 = equity_clean.set_index('Date')
equity_clean_2.head()
```

```python
#Resetting the Dataframe with the dropped column
new_first_equity = equity_clean_2.reset_index()
new_first_equity
```

```python
#Read In Second Equity Dataframe Prices
second_equity_rtns = pd.read_csv('final_equity_returns.csv')
second_equity_rtns.head()
```

```python
#Read In Fixed Income Dataframe Prices
fixed_income_rtns = pd.read_csv('final_fixed_income.csv')
fixed_income_rtns.head()
```

```python
#Dropping the First Row in the Fixed Income DataFrame
fixed_income_final = fixed_income_rtns.drop([0])
fixed_income_final
```

```python
#Setting Date Column = Date
fixed_income_clean = fixed_income_final.set_index('Date')
fixed_income_clean.head()
```

```python
#Resetting the Dataframe with the dropped column
new_fixed_income = fixed_income_clean.reset_index()
new_fixed_income
```

```python
#Dropping the Date Column in Fixed Income DataFrame 
new_fixed_income_1 = new_fixed_income.drop(columns = ['Date'])
new_fixed_income_1
```

```python
#Concat all DataFrames for Full Universe
rtns_universe = pd.concat([new_first_equity,second_equity_rtns,new_fixed_income_1],axis = 'columns', join = 'inner')
rtns_universe
```

```python
#Setting the Blended Optimized ETFs
blended_universe = rtns_universe[['SMH', 'XLU','QQQ','TLT']]
blended_universe
```

```python
#Percent Change for the Blended Optimized Universe
blended_returns = blended_universe.pct_change()
blended_returns
```

```python
# Calculate weighted blended returns 
blended_portfolio_returns = blended_returns.dot(blend_opt_weights)
blended_portfolio_returns.head()
```

```python
#Calculate the cumlative returns 
cumlative_blended_port_returns = (1 + blended_portfolio_returns).cumprod()
cumlative_blended_port_returns.head()
```

```python
#Setting the Blended Port STD
blended_port_std = blended_portfolio_returns.std()
blended_port_std
```

```python
# Rolling 60 Day Standard Deviation 
rolling_blended_std = blended_portfolio_returns.rolling(window=60).std()
rolling_blended_std
```

```python
#Setting the Equity Optimized ETFs 
equity_opt_universe = rtns_universe[['SMH', 'XLU','QQQ']]
equity_opt_universe
```

```python
#Setting the returns for equity optimization
equity_opt_rtns = equity_opt_universe.pct_change()
equity_opt_rtns
```

```python
# Calculate weighted equity optimized returns 
equity_opt_port_returns = equity_opt_rtns.dot(equity_opt_weights)
equity_opt_port_returns.head()
```

```python
#Calculate the equity optimized cumlative returns 
cumlative_equity_port_returns = (1 + equity_opt_port_returns).cumprod()
cumlative_equity_port_returns.head()
```

```python
#Setting the Equity Optimized Port STD
equity_opt_std = equity_opt_port_returns.std()
equity_opt_std
```

```python
# Rolling 60 Day Standard Deviation 
rolling_blended_std = blended_portfolio_returns.rolling(window=60).std()
rolling_blended_std
```

```python
#Setting the Equity Balanced Optimized ETFs 
equity_equal_universe = rtns_universe[['VOOV', 'EZU','IWM', 'SMH', 'IBB', 'XLU', 'VOOG', 'SPY', 'QQQ', 'BKF', 'EEM', 'AIA']]
equity_equal_universe
```

```python
#Setting the returns for Equity Balaced
equity_equal_rtns = equity_equal_universe.pct_change()
equity_equal_rtns
```

```python
# Calculate weighted equity optimized returns 
equity_equal_port_rtns = equity_equal_rtns.dot(equity_equal_weight_weights)
equity_equal_port_rtns.head()
```

```python
#Setting the Standard Deviation for the Equity Balanced Portfolio
equity_equal_std = equity_equal_port_rtns.std()
equity_equal_std
```

```python
# Rolling 60 Day Standard Deviation 
rolling_equity_equal_std = equity_equal_port_rtns.rolling(window=60).std()
rolling_equity_equal_std
```

```python
#Setting the Fixed Income Optimized ETFs 
fixed_opt_universe = rtns_universe[['TLT']]
fixed_opt_universe
```

```python
#Setting the returns for the Fixed Optimized Portfolio
fixed_opt_rtns = fixed_opt_universe.pct_change()
fixed_opt_rtns
```

```python
#Setting the Standard Deviation for Fixed Optimized Portfolio
fixed_opt_std = fixed_opt_rtns.std()
fixed_opt_std
```

```python
# Rolling 60 Day Standard Deviation 
rolling_fixed_opt_std = fixed_opt_rtns.rolling(window=60).std()
rolling_fixed_opt_std
```

```python
#Setting the Fixed Income Equal ETFs
fixed_equal_universe = rtns_universe[['AGG', 'TLT','SHY','IEI', 'HYG', 'BKLN', 'LQD']]
fixed_equal_universe
```

```python
#Setting the returns for the fixed income equal universe
fixed_equal_rtns = fixed_equal_universe.pct_change()
fixed_equal_rtns
```

```python
# Calculate weighted fixed equal returns 
fixed_equal_port_rtns = fixed_equal_rtns.dot(fixed_income_equal_weights)
fixed_equal_port_rtns.head()
```

```python
# Setting the Standard Deviation for the Fixed Equal Portfolio
fixed_equal_std = fixed_equal_port_rtns.std()
fixed_equal_std
```

```python
#Setting the rolling 60day STD for the fixed equal portfolio
rolling_fixed_equal_std = fixed_equal_port_rtns.rolling(window=60).std()
rolling_fixed_equal_std
```
